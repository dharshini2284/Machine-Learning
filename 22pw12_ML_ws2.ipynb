```python
#importing rquired packages
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#defining the datas as a array
x=np.array([2,3,5,7,9])
y=np.array([4,5,7,10,15])

#finding slope and intercept using polyfit to form a equation
m,c = np.polyfit(x,y,1)
print("The linear regression equation is y = ",round(m,2),"*x + ",round(c,2))

#finding correlation coefficient using corrcoef
correlation=np.corrcoef(x,y)
print("The correlation coefficient of two variables is ",round(correlation[0,1],3))

#finding the residual to calculate error
y_predicted = m*x + c
residual = np.sum((y-y_predicted)**2)
print("The least squares of errors is ",round(residual,2))

#plotting the given data as box plot without outliers
plt.boxplot(y)
plt.title("Given data as box plot without outliers")
plt.xlabel("y")
plt.show()

#defining array by adding a extra element to witness a outlier
X=np.array([2,3,5,7,9,64])
Y=np.array([4,5,7,10,15,52])

#finding the slope and intercept for the defined array
M,C=np.polyfit(X,Y,1)
print("The linear regression equation is y = ",round(M,2),"*x + ",round(C,2))

#finding the correlation coefficient using corrcoef
correlation1=np.corrcoef(X,Y)
print("The correlation coefficient of two variables is ",round(correlation1[0,1],3))

#finding the residual to calculate error
y_predicted1=M*X+C
residual1=np.sum((Y-y_predicted1)**2)
print("The least squares of errors is ",round(residual1,2))

#plotting the data as box plot with outliers
plt.boxplot(Y)
plt.title("Given data as box plot with outliers")
plt.xlabel("Y")
plt.show()
```

```python

#importing required packages
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#defining the given data as array
x=np.array([14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2])
y=np.array([215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408])

#finding the slope and intercept using polyfit
m,c=np.polyfit(x,y,1)
y_pred=m*x+c
print(f"The linear regression is y=",round(m,2),"*x + ",round(c,2))

#plotting the data points as scatter plot
plt.scatter(x,y,color="black",label="Data points")

#plotting the best fit line
plt.plot(x,y_pred,color="red",label="Line of Best Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

#finding interpolate sales at 21
x_inter=21
x_interpolate=m*x_inter+c
print("Interpolated sales value at 21°: ",round(x_interpolate,2))

#finding extrapolate sales at 29
y_inter=29
y_extrapolate=m*y_inter+c
print("Extrapolated sales value at 29°: ",round(y_extrapolate,2))
```

```python
#importing required packages
import pandas as pd
import statsmodels.api as sm

#loading the dataset
data = pd.read_csv('/content/sample_data/advertising.csv')

#defining predictor (X) and response (y) variables
X = data['TV']
Y = data['Sales']

#adding constant to the predictor for intercept calculation
X = sm.add_constant(X)

#fit the linear regression model
model = sm.OLS(Y, X).fit()

#printing the model summary
print(model.summary())

#calculating confidence interval around the slope
confidence_interval = model.conf_int(alpha=0.05)
print("\n95% Confidence Interval for the slope of TV:\n",confidence_interval)
```
