```python
#Import the required packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Total number of customers
n = 20
# Probability of success
p = 0.1
# Expected number of successful trials
k = 5

# Calculates the probability of exactly 5 customers returning items using the binomial probability mass function (PMF)
prob_5_returns = binom.pmf(k, n, p)
print(f"Probability that exactly 5 customers will return the items: {prob_5_returns:.4f}")

# Creates an array of possible outcomes (0 to n)
x = np.arange(0, n + 1)

# Calculates the PMF values for all possible outcomes
pmf_values = binom.pmf(x, n, p)
print("PMF values:")
for i, pmf in enumerate(pmf_values):
    print(f"P(X = {i}) = {pmf:.4f}")

# Generates and show the plot of the binomial PMF
plt.bar(x, pmf_values, color='blue', alpha=0.7)
plt.xlabel("Number of returns")
plt.ylabel("Probability")
plt.title("Binomial PMF - Number of returns in 20 customers")
plt.show()
```

```python
#Import the required packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Total number of customers
n = 20
# Probability of success
p = 0.1

# Calculates the cumulative probability of at most 5 customers returning items using the binomial CDF
max_5_returns = binom.cdf(5, n, p)
print(f"Probability of a maximum of 5 customers returning the items: {max_5_returns:.4f}")

# Creates an array of possible outcomes (0 to n)
x = np.arange(0, n + 1)

# Calculates the PMF values for all possible outcomes
pmf_values = binom.pmf(x, n, p)

# Plots the PMF values
plt.bar(x, pmf_values, color="green", alpha=0.7)
plt.xlabel("Number of returns")
plt.ylabel("Probability")
plt.title("Binomial PMF - Number of returns in 20 customers")
plt.show()
```

```python
#Importing the required packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Total number of customers
n = 20
# Probability of success
p = 0.1

# Calculates the probability of more than 5 customers returning items
# This is 1 minus the cumulative probability of at most 5 customers returning items
more_than_5_returns = 1 - binom.cdf(5, n, p)
print(f"Probability that more than 5 customers will return the items purchased by them: {more_than_5_returns:.4f}")

# Creates an array of possible outcomes (0 to n)
x = np.arange(0, n + 1)

# Calculates the PMF values for all possible outcomes
pmf_values = binom.pmf(x, n, p)

# Plots the PMF values
plt.bar(x, pmf_values, color="green", alpha=0.7)
plt.xlabel("Number of returns")
plt.ylabel("Probability")
plt.title("Binomial PMF - Number of returns in 20 customers")
plt.show()
```

```python
#Importing the required packages
from scipy.stats import binom
import numpy as np

# Total number of customers
n = 20
# Probability of success
p = 0.1

# Calculates the mean and variance of the binomial distribution
mean, var = binom.stats(n, p)

# Calculates the standard deviation from the variance
std_dev = np.sqrt(var)

# Printing the mean, variance, and standard deviation
print("The Mean is", mean)
print("The Variance is", var)
print("The Standard Deviation is", std_dev)
```

```python
#Importing the required packages
from scipy.stats import poisson

# Average number of calls per hour
lambda_hour = 10

# Calculates the cumulative probability of receiving at most 5 calls in an hour using the Poisson CDF
prob_5_max = poisson.cdf(5, lambda_hour)

# Calculates the probability of receiving more than 30 calls in a 3-hour period
lambda_3_hours = 3 * lambda_hour
prob_30_max = 1 - poisson.cdf(30, lambda_3_hours)

# Printing the probabilities
print("Probability that the number of calls will be at most 5:", prob_5_max)
print("Probability that the number of calls over a 3-hour period will exceed 30:", prob_30_max)
```

```python
#Importing the required packages
import numpy as np

# Mean Time Between Failures (MTBF) in hours
MTBF = 1000

# Calculates the probability that the system will fail before 1000 hours
prob_fail_before_1000 = 1 - np.exp(-1000 / MTBF)
print(f"Probability that the system will fail before 1000 hours: {prob_fail_before_1000:.4f}")

# Calculates the probability that the system will survive up to 2000 hours
prob_survive_up_to_2000 = np.exp(-2000 / MTBF)
print(f"Probability that the system will survive up to 2000 hours: {prob_survive_up_to_2000:.4f}")

# Calculates the time by which 10% of the systems will fail (P10 life)
p10_life = -MTBF * np.log(0.90)
print(f"Time by which 10% of the systems will fail (P10 life): {p10_life:.2f} hours")
```

```python
#Importing the required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Loading the CSV file from the specified URL
url = '/content/sample_data/BEML.csv'
df = pd.read_csv(url)

# Extracting the 'Close' column from the dataframe
beml_close = df['Close']

# Plots the 'Close' prices
plt.plot(beml_close, label="BEML Close Price", color="pink")
plt.title("BEML Stock Close Price Trend")
plt.xlabel("Days")
plt.ylabel("Close Price")
plt.legend()
plt.show()
```

```python
gain=beml_close.pct_change().dropna()
plt.plot(gain,label="BEML Daily Returns(gain)",color="green")
plt.title("BEML Daily returns")
plt.xlabel("Days")
plt.ylabel("Daily Returns")
plt.show()

#gain chart
df['Date']=pd.to_datetime(df['Date'])
df['Daily Return']=df['Close'].pct_change().dropna()
plt.plot(df['Date'][1:],df['Daily Return'][1:],label="BEML Daily Returns")
plt.title("BEML Daily Returns")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.show()

#check it is normally distributed
import seaborn as sns
sns.histplot(df['Daily Return'].dropna(),kde=True,label="BEML Daily Return")
plt.title("Histogram of BEML Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.show()

#Shapiro-Wilk test for normality
from scipy.stats import shapiro,norm
stat,p_value=shapiro(df['Daily Return'].dropna())
print(f'Shapiro-Wilk Test:W={stat},p-value={p_value}')
if p_value>0.05:
   print('BEML daily returns are normally distributed.')
else:
    print('BEML daily returns are not normally distributed.')

#expected daily return
expected_return=df['Daily Return'].mean()
print(f'Expected Daily Return:{expected_return:.4f}')

#volatility
volatility=df['Daily Return'].std()
print(f'Volatility:{volatility:.6f}')

confidence_interval=norm.interval(0.95,loc=expected_return,scale=volatility)
print(f'95% Confidence Interval for Daily Returns: {confidence_interval}')
```

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Generating random data
np.random.seed(0)  # For reproducibility
x = np.array(7 * np.random.rand(100, 1) - 3)  # Random values for x in range [-3, 4]
y = 13 * x**2 + 2 * x + 7  # Generating y based on the given quadratic function

# (I) Linear Regression

# Fitting linear regression model
linear_model = LinearRegression()
linear_model.fit(x, y)

# Predicting with linear model
y_pred_linear = linear_model.predict(x)

# Calculating performance metrics
mse_linear = mean_squared_error(y, y_pred_linear)
rmse_linear = np.sqrt(mse_linear)
r2_linear = r2_score(y, y_pred_linear)

print("Linear Regression Performance:")
print(f"Mean Squared Error: {mse_linear:.2f}")
print(f"Root Mean Squared Error: {rmse_linear:.2f}")
print(f"R^2 Score: {r2_linear:.2f}")

# Plotting data points and linear line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_pred_linear, color='red', label='Linear fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()

# (II) Polynomial Regression with degree 2

# Transforming data to include polynomial features
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_features.fit_transform(x)

# Fitting polynomial regression model
poly_model = LinearRegression()
poly_model.fit(x_poly, y)

# Predicting with polynomial model
y_pred_poly = poly_model.predict(x_poly)

# Calculating performance metrics
mse_poly = mean_squared_error(y, y_pred_poly)
rmse_poly = np.sqrt(mse_poly)
r2_poly = r2_score(y, y_pred_poly)

print("\nPolynomial Regression Performance:")
print(f"Mean Squared Error: {mse_poly:.2f}")
print(f"Root Mean Squared Error: {rmse_poly:.2f}")
print(f"R^2 Score: {r2_poly:.2f}")

# Plotting quadratic equation
x_range = np.linspace(-3, 4, 100).reshape(-1, 1)
x_range_poly = poly_features.transform(x_range)
y_range_poly = poly_model.predict(x_range_poly)

plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x_range, y_range_poly, color='green', label='Polynomial fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.show()

# Comparison
print("\nComparison:")
print(f"Linear Regression vs Polynomial Regression:")
print(f"Linear Regression R^2 Score: {r2_linear:.2f}")
print(f"Polynomial Regression R^2 Score: {r2_poly:.2f}")
```
